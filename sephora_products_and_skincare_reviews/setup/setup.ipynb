```python
%pip install kagglehub -q
dbutils.library.restartPython()

import kagglehub
import os
from pathlib import Path

spark.sql("CREATE VOLUME workspace.sephora_products_and_skincare_reviews.data;")

path = Path("/Volumes/workspace/sephora_products_and_skincare_reviews/data") / "source"
os.makedirs(path, exist_ok=True)

kagglehub.dataset_download(
  "nadyinky/sephora-products-and-skincare-reviews",
  output_dir=path
)
```

```python
%sql
CREATE OR REPLACE TABLE workspace.sephora_products_and_skincare_reviews.bronze_products_info AS
  SELECT
    *,
    _metadata.file_path AS source_file
  FROM read_files(
    '/Volumes/workspace/sephora_products_and_skincare_reviews/data/source/product_info.csv',
    format => 'CSV',
    inferSchema => true,
    header => true
);
```

```python
%sql
SELECT *
FROM workspace.sephora_products_and_skincare_reviews.bronze_products_info
LIMIT 10;
```

```python
%sql
CREATE OR REPLACE TABLE workspace.sephora_products_and_skincare_reviews.bronze_reviews AS
  SELECT
    *,
    _metadata.file_path AS source_file
  FROM read_files(
    '/Volumes/workspace/sephora_products_and_skincare_reviews/data/source/reviews_*.csv',
    format => 'CSV',
    inferSchema => true,
    header => true
);
```

```python
%sql
SELECT *
FROM workspace.sephora_products_and_skincare_reviews.bronze_reviews
LIMIT 10;
```

```python
%sql
SELECT DISTINCT source_file
FROM workspace.sephora_products_and_skincare_reviews.bronze_reviews;
```
